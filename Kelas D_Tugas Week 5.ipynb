1. Buatlah program pemanggilan angka secara descending menggunakan pengulangan dari angka 50 hingga 0!

```python
#membuat program pemanggilan angka secara descending pengulangan angka 50 hingga 0
no = 50
for i in range(0, no+1):
    for k in range(0, no+1):
        print(k, end='')
print()
no-=1
```

Buatlah sebuah list yang berisi kota-kota yang ada di indonesia (minimal 10, salah satunya kota asal kalian), kemudian buat pengulangan yang menampilkan isi-isi dalam list tersebut, jika salah satu kota menunjukkan kota asal kalian maka tambahkan output "kota X adalah kampung halaman saya" (program tetap lanjut)

```python
#buat list berisi kota-kota di indonesia 
kota = ['Tangerang', 'Jakarta', 'Medan', 'Bogor', 'Depok', 'Bekasi', 'Manado', 'Padang', 'Jogja', 'Palembang']

for k in kota:
    print(k)

    if k == kota[0]:
        print('Kota Tangerang adalah kampung halaman saya')
```

Abdul adalah seorang mahasiswa semester 2. Selama satu minggu ia membuat daftar menu sahur yang dia makan selama seminggu yang lalu. dengan catatan hari kamis Abdul ketiduran sehingga dia tidak sahur. Berdasarkan kasus tersebut buatlah program pengulangannya!

```python
#buat program pengulangannya
hari = ['senin', 'selasa','rabu','kamis','jumat','sabtu', 'minggu']
menu_sahur = ['soto','pecel lele', 'ayam goreng', 'sayur sop', 'rendang','telur balado']

for i in range((len(hari))-1):
    if hari[i] == 'kamis' :
        print(hari[i] , 'Abdul tidak sahur karena ketiduran')
    else :
        print(hari[i] , ':', menu_sahur[i])
```
